**Step 1 - Robot resizable!**

Let's get started! In the block below is Dr. Celine's robot head from Week 2 (but using variables, not constants, because we need these to be variables later). Run this to see what the robot looks like. Change the canvas size() and run it again to make sure the robot resizes appropriately.  

```java
/** 
Dr. Celine's Robot Head
**/
// global vars
size(300,300);
int headWidth = width*2/3;
int headHeight = height/3;
int headTop = (height - headHeight)/3;
int headLeft = (width - headWidth)/2;

int leftEyeX = headLeft + headWidth/4;
int rightEyeX = headLeft + headWidth*3/4;
int eyeY = headTop + headHeight/3;
int eyeDiam = headWidth/5;
int eyeBallDiam = eyeDiam/2;

int mouthLeft = leftEyeX;
int mouthRight = rightEyeX;
int mouthY = headTop + headHeight*3/4;


background(100, 100, 250);

// draw head
fill(200, 200, 50);
rect(headLeft, headTop, headWidth, headHeight);

// draw eyes
fill(255);
ellipse(leftEyeX, eyeY, eyeDiam, eyeDiam); // left eye
ellipse(rightEyeX, eyeY, eyeDiam, eyeDiam); // right eye

// draw eyeballs
fill(0);
ellipse(leftEyeX, eyeY, eyeBallDiam, eyeBallDiam); // left eye
ellipse(rightEyeX, eyeY, eyeBallDiam, eyeBallDiam); // right eye


// draw mouth
stroke(200, 100, 100);
strokeWeight(8);
line(mouthLeft, mouthY, mouthRight, mouthY);
```
