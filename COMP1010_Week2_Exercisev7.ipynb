**Step 8 - CONSTANTS!**

Okay, so in actual fact, all of the things we have created so far are CONSTANTS, not variables. How do you know? Well, because each of them gets assigned only once during the program and the values don't change while the program is running. 

So, I hate to do this to you, but I want you to copy your program above, and add the keyword final in front of all your declarations, and change the code to be the correct style (ALL_CAPS). As an example, these:

    int headHeight = height/3;

    int headWidth = width\*2/3;

    int headTop = (height - headHeight)/3;

    int headLeft = (width - headWidth)/2;

should be replaced by these:

    final int HEAD_HEIGHT = height/3;

    final int HEAD_WIDTH = width\*2/3;

    final int HEAD_TOP = (height - HEAD_HEIGHT)/3;

    final int HEAD_LEFT = (width - HEAD_WIDTH)/2;
