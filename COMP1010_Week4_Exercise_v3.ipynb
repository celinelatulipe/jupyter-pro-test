**Step 1 - Random circles**

Let's get started! The script below is a basic template for an active Processing script. The objective is to make a program that draws randomly colored dots of random diameters, on the screen and let those colored dots build up over time. There is no user interaction in this. However, we don't want the dots to be too small, and we don't want them to be too large, so we are going to set maximum and minimun sizes of the dots using global constants. We also want to make sure the colors are all bright, so we are doing to set maximum and minimum color ranges. Follow this sub-steps below to do this.

Declare four global constants for MAX_DOT_SIZE, MIN_DOT_SIZE, MIN_COLOR, MAX_COLOR. MAX_COLOR should be 255, MIN_COLOR should be 100. You can decide on the min and max dot size. 

**Step 1b**
Add a fill command to draw() to set the fill colour. You want to call random() for each of the three colors. Use the two constants you created to give a range to random.

**Step 1c**
Add an ellipse command to draw a dot. Use random() for the x and y coordinates, so that the dots can go anywhere on the canvas. To make circular dots, you need the width and height of the ellipse to be the same. So, you will need to create a local variable above the ellipse command to randomly generate a size. Remember that random() returns a float, so your local variable will have to be a float. Use the two constants you created (MIN_DOT_SIZE and MAX_DOT_SIZE) as the range for generating a random size.

```java
/** 
Basic Active Processing Script
**/
final int MIN_DOT_SIZE = 20;
final int MAX_DOT_SIZE = 40;
final int MIN_COLOUR = 100;
final int MAX_COLOUR = 256;

void setup() {
    size(500, 500);
    background(0);
}

void draw() {
    fill(random(MIN_COLOUR, MAX_COLOUR), random(MIN_COLOUR, MAX_COLOUR), random(MIN_COLOUR, MAX_COLOUR));
    float dotSize = random(MIN_DOT_SIZE, MAX_DOT_SIZE);
    noStroke();
    ellipse(random(width), random(height), dotSize, dotSize);
}
```

**Step 2d**
Constrain the random walk so that it stays on the canvas. After adding a random amount to dotX and dotY, use min() and max() functions so that dotX is either 0, dotX or width. Then use min() and max() to set dotY, so that it is either 0, dotY, or height. These calls will ensure the dots are not going to walk off the canvas.

```java
// copy the code from above and then make the dots do a random walk around the screen
/** 
Basic Active Processing Script
**/
final int MIN_DOT_SIZE = 20;
final int MAX_DOT_SIZE = 40;
final int MIN_COLOUR = 100;
final int MAX_COLOUR = 256;
int dotX;
int dotY;
final int MAX_MOVE = 10;

void setup() {
    size(500, 500);
    dotX = width/2;
    dotY = height/2;
    background(0);
}

void draw() {
    fill(0, 10);
    rect(0,0, width, height);
    fill(random(MIN_COLOUR, MAX_COLOUR), random(MIN_COLOUR, MAX_COLOUR), random(MIN_COLOUR, MAX_COLOUR));
    float dotSize = random(MIN_DOT_SIZE, MAX_DOT_SIZE);
    noStroke();
    dotX = dotX + (int)random(MAX_MOVE*-1, MAX_MOVE);
    dotY = dotY + (int)random(MAX_MOVE*-1, MAX_MOVE);
    dotX = max(dotX, 0);
    dotX = min(dotX, width);
    dotY = max(dotY, 0);
    dotY = min(dotY, height);
    ellipse(dotX, dotY, dotSize, dotSize);
}
```
