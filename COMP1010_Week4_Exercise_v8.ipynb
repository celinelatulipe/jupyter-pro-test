**Step 6 - BONUS - Make a moon orbit around the earth, which is orbiting the sun**

The code below generates an abstract space scene in which earth is orbiting the sun, while stars flash by in the background. Run the code to see it work, and then your job is to add constants and variables to make a moon orbit the earth. You will want to inspect the code to understand how it works. Note that we use the trig functions sin() and cos() to calculate the position of the earth with respect to the moon, using the EARTH_DIST constant, to keep the orbit the same all the way around. Note how the speed of the orbit is controlled by EARTH_SPEED (which just slighlty increments the orbit angle during each call to draw()). 

To complete this bonus, you need to make a function that draws a white moon that is smaller in size than the earth, and that orbits the earth (the way the earth orbits the sun). The moon should orbit the earth faster than the earth orbits the sun. The moon should be close to the earth. Make sure to create constants and variables that have good names.

Your moon orbiting the earth should look something like this:

<img src="images/moon.png" width=300 />

```java
final int CENTERX = 250;
final int CENTERY = 250;
final int SUN_DIAM = 70;
final int EARTH_DIAM = 50;
final int EARTH_DIST = 150; // distance from sun
final float EARTH_SPEED = 0.01;
float earthAngle = 0.0; // angle of earth from sun
float earthX, earthY;

void setup() {
  size(500,500);

}

void draw() {
  background(0);
  
  drawStars();
  drawSun();
  drawEarth();
  
}

void drawStars() {
  stroke(255); // white
  strokeWeight(3);
  point(random(width), random(height));
  point(random(width), random(height));
  point(random(width), random(height));
  point(random(width), random(height));
}

void drawSun() {
  fill(255, 200, 100); // yellowish
  noStroke();
  ellipse(width/2, height/2, SUN_DIAM, SUN_DIAM);
}

void drawEarth() {
  fill(100, 100, 250); // blueish
  earthX = CENTERX + cos(earthAngle)*EARTH_DIST;
  earthY = CENTERY + sin(earthAngle)*EARTH_DIST;
  ellipse(earthX, earthY, EARTH_DIAM, EARTH_DIAM);
  earthAngle = earthAngle + EARTH_SPEED;
}
```
